### Step 0

Run the necessary imports.

```python
import cv2
import numpy as np
import random
import dlib
from matplotlib import pyplot as plt
plt.rcParams['figure.figsize'] = [15, 10]
```

```python
def generate_random_color():
    return (random.randint(0,256),random.randint(0,256),random.randint(0,256))

def rect_to_bb(rect):
    # Dlib rect --> OpenCV rect
    x = rect.left()
    y = rect.top()
    w = rect.right() - x
    h = rect.bottom() - y

    return (x, y, w, h)
```

```python
generate_random_color()
```

### Step 1

Load an image (any image that contains faces).

```python
def get_image(path):
    img = cv2.imread(path)
    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    return img

def load_image(img):
    img = cv2.imread('../data/normal_photo.jpeg')
    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    gray = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
    plt.imshow(img)
    return gray
```

### Step 2

Load the dlib face predictor.

```python
def load_dlib_face_predictor(gray):
    # Let's load the detector
    detector = dlib.get_frontal_face_detector()
    # Detect faces, see http://dlib.net/face_detector.py.html
    # 1 --> upsampling factor
    rects = detector(gray, 1)
    print('Number of detected faces:', len(rects))
    print(rects)
    print(rects[0].left)
    
    return rects
```

### Step 3

Run the predictor on your image.

```python
generate_random_color()
```

```python
def draw_rectangles(rects, img):

    # Draw rectangle around each face
    result_dlib = np.copy(img)
    faces_dlib_img = []

    for rect in rects:    
        # Draw rectangle around the face
        x1, y1, w, h = rect_to_bb(rect)
        x2 = x1 + w
        y2 = y1 + h

        print(x1, y1, w, h)
        color = generate_random_color()
        print(color)

        cv2.rectangle(result_dlib, (x1, y1), (x2, y2), color, 17)
        faces_dlib_img.append(img[y1:y2, x1:x2, :])
    

    plt.subplot(122), plt.imshow(result_dlib), plt.title('dlib')
    
def detect_faces(path):
    img = get_image(path)
    gray = load_image(img)
    rects = load_dlib_face_predictor(gray)
    draw_rectangles(rects, img)
    
```

### Step 4

Draw bounding boxes around the detected faces and plot the image. Use different colour for each face.

```python
detect_faces('../data/normal_photo.jpeg')
```
